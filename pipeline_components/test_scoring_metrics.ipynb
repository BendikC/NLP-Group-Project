```python
# Example data
predictions = [
        "the quick brown fox",
        "jumps over the lazy dog",
        "this is an example",
        "prediction is not the same as truth",
        "another example for testing",
        "semantic similarity matters",
        "short sentence",
        "a very long sentence that keeps going on and on to test the limits of evaluation metrics",
        "common phrase repetition repetition",
        "edge cases are always tricky"
    ]

truths = [
        "the quick fox",
        "jumps over the lazy dog",
        "this is a different example",
        "truth is not prediction",
        "another example test case",
        "semantic similarity counts",
        "a short sentence",
        "a very long sentence that keeps going on and on testing the boundaries of evaluation",
        "common phrase repetition",
        "edge cases are tricky indeed"
    ]
```

```python
# Test Exact Match
em_score = ScoringMetrics.exact_match(predictions, truths)
print(f"Exact Match Score: {em_score}")
```

```python
# Test Cover Exact Match
cem_score = ScoringMetrics.cover_exact_match(predictions, truths)
print(f"Cover Exact Match Score: {cem_score}")
```

```python
# Test F1 Score
f1 = ScoringMetrics.f1_score(predictions, truths)
print(f"F1 Score: {f1}")
```

```python
# Test BLEU
bleu = ScoringMetrics.bleu(predictions, truths)
print(f"BLEU Score: {bleu}")
```

```python
# Test METEOR
meteor = ScoringMetrics.meteor(predictions, truths)
print(f"METEOR Score: {meteor}")
```
